# Найти максимальный срез в списке с изменяющимися знаками элементов. 0 можно считать как + так и -.

```python
def solution(A):
    best = 1
    i = 0
    counter = 1
    while 1:
        counter = 1
        for j in range(i+1, len(A)):
            if A[j] * A[j-1] > 0:
                break
            elif A[j] * A[j-1] < 0:
                counter += 1
            elif A[j] == 0:
                counter += 1
            elif A[j-1] == 0:
                if counter == 1:
                    counter += 1
                elif counter > 1:
                    if A[j] * A[j-2] >= 0:
                        counter += 1
                    else:
                        break



        if counter > best:
            best = counter
        i = j

        if i == len(A) -1:
            break

    return best
```

```python
A = [5, 4, -3, 2, 0, 1, -1, 0, 2, -3, 4, -5]
solution(A)
```
